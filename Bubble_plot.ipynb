```python
import pandas as pd

df = pd.read_csv("/content/bubble_5vs5.csv")
df.columns
```

```python
df_sorted = df.sort_values('log2FE', ascending=True)

plt.figure(figsize=(12, 16))
sc = plt.scatter(
    df_sorted['log2FE'],
    df_sorted['Term'],
    s=df_sorted['Count'] * 4,
    c=df_sorted['_log10P'],
    cmap='Purples',
    alpha=0.8,
    edgecolors='black',
    linewidths=0.5
)

cbar = plt.colorbar(sc)
cbar.set_label('-log10(P-value)', rotation=270, labelpad=24)

plt.xlabel('log2 Fold Enrichment')
plt.ylabel('')
plt.title('Biological Process Enrichment', fontsize=16, fontweight='bold')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.tight_layout()
plt.show()
```

```python
df_sorted = df.sort_values('log2FE', ascending=True)

plt.figure(figsize=(20, 16))
sc = plt.scatter(
    df_sorted['log2FE'],
    df_sorted['Term'],
    s=df_sorted['Count'] * 16,
    c=df_sorted['_log10P'],
    cmap='Purples',
    alpha=0.8,
    edgecolors='black',
    linewidths=0.5
)

cbar = plt.colorbar(sc)
cbar.set_label('-log10(P-value)', rotation=270, labelpad=24, fontsize=24)
cbar.ax.tick_params(labelsize=24)  # colorbar ticks



plt.xlabel('log2 Fold Enrichment', fontsize=24)
plt.ylabel('', fontsize=24)
plt.title('Biological Process Enrichment', fontsize=24, fontweight='bold')
plt.xticks(fontsize=24)
plt.yticks(fontsize=30)
plt.tight_layout()
plt.show()
```
